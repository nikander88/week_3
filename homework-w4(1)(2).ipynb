```python
B = pd.read_csv("airport_temp.csv",dayfirst=True,sep=",", 
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])


D = pd.read_csv("lepola_temp.csv",dayfirst=True,sep=",", 
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])

```

**(3) Node $A$ needs to identify risky situations in Lappeenranta when the temperature is above $25$ degrees and the wind is above $5$ m/s based on the sensors' signals. If a risky situation is identified at node $A$ as binary number "1" while normal situations as "0". How would you solve this problem? Plot the output signal from node $A$ and discuss its relation to the input signals from $B$-$E$.**

```python
A = (B['2018-7'].values>25) & (C['2018-7'].values>5) & (D['2018-7'].values>25) & (E['2018-7'].values>5)
plt.figure(figsize=(16,6)) 
plt.plot(A,color='black', marker='.',linestyle='')
plt.title("Shows with binars if the Air temperature in Lappeenranta goes over 25 degrees and the wind speed over 5m/s ") 
```

```python
The nodes B-E are turned into binary, so every row has a true or false 
statement instead of wind speed or temperature. Node A reads all of them 
and if all nodes have true statement, then A has a true statement, which means 
that there is a risky situation. 
```
